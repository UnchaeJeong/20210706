## 단어장 불러오기

```python
from os.path import join
base_dir = '/content/drive/MyDrive/T_academy/PROJECTS/StartupChallenge_Zeroxflow'
```

```python
ele_path = join(base_dir, 'data', 'ele.csv')
mid1_path = join(base_dir, 'data', 'mid1.csv')
mid2_path = join(base_dir, 'data', 'mid2.csv')
mid3_path = join(base_dir, 'data', 'mid3.csv')
high1_path = join(base_dir, 'data', 'high1.csv')
high2_path = join(base_dir, 'data', 'high2.csv')
high3_path = join(base_dir, 'data', 'high3.csv')


ele = pd.read_csv(ele_path, usecols=[0], names=['eng'], encoding='utf-8').dropna(axis=0).reset_index(drop=True)
mid1 = pd.read_csv(mid1_path, usecols=[0], names=['eng'], encoding='utf-8').dropna(axis=0).reset_index(drop=True)
mid2 = pd.read_csv(mid2_path, usecols=[0], names=['eng'], encoding='utf-8').dropna(axis=0).reset_index(drop=True)
mid3 = pd.read_csv(mid3_path, usecols=[0], names=['eng'], encoding='utf-8').dropna(axis=0).reset_index(drop=True)
high1 = pd.read_csv(high1_path, usecols=[0], names=['eng'], encoding='utf-8').dropna(axis=0).reset_index(drop=True)
high2 = pd.read_csv(high2_path, usecols=[0], names=['eng'], encoding='utf-8').dropna(axis=0).reset_index(drop=True)
high3 = pd.read_csv(high3_path, usecols=[0], names=['eng'], encoding='utf-8').dropna(axis=0).reset_index(drop=True)
```

```python
# NA 없다.
display(ele.info(), mid1.info(), mid2.info(), mid3.info(), high1.info(), high2.info(), high3.info())
```

```python
# 잘 들어왔다.
display(ele, mid1, mid2, mid3, high1, high2, high3)
```

## level 칼럼 추가

```python
# level 칼럼 추가

for i,w in enumerate([ele, mid1, mid2, mid3, high1, high2, high3]):
  w['level'] = i
```

```python
display(ele, mid1, mid2, mid3, high1, high2, high3)
```

## concat

```python
# concat
words = pd.concat([ele, mid1, mid2, mid3, high1, high2, high3]).reset_index(drop=True)
words
```

```python
ele.shape[0] + mid1.shape[0] + mid2.shape[0] + mid3.shape[0] + high1.shape[0] + high2.shape[0] + high3.shape[0]
```

```python
words.duplicated(subset='eng').sum() # 642개 겹침
```

```python
# 중복 제거
words.drop_duplicates(subset='eng', ignore_index=True, inplace=True)
```

```python
words.shape
```

```python
words.groupby('level').describe()
```

```python
# 어디에 중복이 가장 많았나? 중1
print(ele.shape[0], mid1.shape[0], mid2.shape[0], mid3.shape[0], high1.shape[0], high2.shape[0], high3.shape[0])
```

```python
# savepoint

words_path = join(base_dir, 'data', 'concat_words.csv')
words.to_csv(words_path, index=False, encoding='utf-8')
```

## 자잘한 다듬기

- 한글 떨구기
- 괄호와 그 안에 있는 내용 제거
- 특수기호 제거(-은 안 빼야?)


```python
import re
```

```python
# 작업용으로 카피
words1 = words.copy(deep=True)
```

```python
# 한글 제거
words1['eng'] = words1['eng'].apply(lambda x: re.compile('[가-힣|ㄱ-ㅎ|ㅏ-ㅣ]+').sub('',x))
```

```python
# 잘 떨궈짐.
words1.iloc[785:795, :]
```

(~~) 제거

```python
words1.loc[1452,'eng']
```

```python
re.compile("\(.*?\)").findall(words1.loc[1452,'eng']) # 잘 되는지 실험
```

```python
# (~~) 제거

words1['eng'] = words1['eng'].apply(lambda x: re.compile("\(.*?\)").sub('',x))
words1.loc[1452,'eng']
```

```python
# 특수기호
words1['eng'] = words1['eng'].apply(lambda x: re.compile('[-=+,#/\?:^$.@*\"※~&%ㆍ!』\\’‘|\(\)\[\]\<\>`\'…》]+').sub('',x))
```

```python
words1.loc[1801,'eng'] 
```

```python
# savepoint

words_path = join(base_dir, 'data', 'concat_words2.csv')
words1.to_csv(words_path, index=False, encoding='utf-8')
```
