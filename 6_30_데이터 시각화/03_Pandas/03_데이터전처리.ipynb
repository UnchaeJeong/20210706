### inplace
- 제공되는 전처리 관련 함수들은 inplace 매개변수를 제공하는 것들이 있다.
- inplace가 없는 함수들은 원본을 수정하거나 원본을 유지한 채 작업이 반영된 새로운 것을 생성해 반환한다.
- inplace를 제공할 경우  True를 넣어주면 원본이 변경되고 False를 넣어주면 원본은 유지하고 작업이 반영된 새로운 것을 생성해 반환한다.
- (inplace가 있는지 꼭 확인해야 함.)

### 행, 열 순서 변경

```python
# 순서 변경함수 3가지.

# 열 순서 변경 - 지정된 컬럼 데이터를 지정된 순서로 가져온다. 
# 원본은 유지하고 순서가 변경된 데이터 프레임을 새롭게 생성해서 반환한다.

a1 = ['과학', '수학', '영어', '국어','성별', '학년', '이름']
df2 = df1.reindex(columns=a1)
display(df1)
display(df2)

# 원본 유지하는 지 확인하기 위해 꼭 2개 다 display해볼 것.
```

```python
# 일부만 작성하면 일부만 추출하는 효과를 얻을 수 있다.
# reindex의 의미는 재정렬의 의미를 가지고 있지만 일부 추출도 가능하다.

df3 = df1.reindex(columns = ['국어', '영어', '수학'])
df3
```

```python
# 일부만 가져온다.
df3 = df1.filter(['국어', '영어', '수학'])
df3
```

```python
# 없는 이름
# filter는 없는 이름을 넣어주면 무시한다.

df4 = df1.filter(['국어', '영어', '한국사'])
df4

#없는 이름 컬럼은 그냥 무시됨. 필터 예 - 코인오락기동전
```

```python
# 행 정렬
# axis : 하나의 함수를 통해 행과 열에 대한 작업을 하는 함수인 경우 작업의 기준을 설정하기 위해 axis라는 매개변수를 제공한다.
# 0이면 행이 기준, 1이면 열이 기준.

# 함수를 보게 되면 inplace, axis(axis가 없다면 행, 컬럼 하나만 작업가능 or 컬럼별 따로 되어있거나) 있는지 확인해볼 것.
# () => [shift] + [tab]

df2 = df1.filter([4,1,3,0,2], axis = 0)
df2
```

```python
# 행 추출
df3 = df1.filter([3,0,1], axis=0)
df3
```

```python
# 없는 인덱스를 지정한다.
# 없으면 추가 reindex, 없으면 무시 axis.

df4 = df1.filter([0, 1, 10], axis=0)
df4
```

### 조건에 해당하는 데이터를 검색해온다.

```python
# 데이터 프레임도 True나 False로 구성된 요소를 넣어주면 True에 해당하는 행들만 가져올 수 있다.
df1 = pd.read_csv('data/grade.csv', encoding='euc-kr')
df1

#원래는 데이터 프레임[] 방식으로 접근을 하면 컬럼을 선택하는 것이지만 True, False로 구성된 것을 넣어주면
# True로 해당하는 행만 가져온다.
df1[[True, False, True, False, True]] 
```

```python
df2 = df1[df1['국어']>90] # 3g 내 각각 연산 후 True에 해당하는 것만 골라서 나옴.
df2
```

```python
# ****검색(and)****
# ***국어 점수가 90점 보다 큰, 수학점수가 80보다 큰 학생 정보를 가져온다. ****
df2 = df1.query('국어 > 90')
df2
```

```python
# 성별이 남자인 학생들
# 식 안에 따옴표가 없는 문자열은 컬럼이름으로 인식하고 따옴표가 있는 것은 문자열 값으로 인식한다.

df2 = df1.query('성별 == "남자"')
df2
```

```python
# ****검색(and)****
# ***국어 점수가 90점 보다 큰, 수학점수가 80보다 큰 학생 정보를 가져온다. ****
df2 = df1.query('국어 > 90 and 수학 > 80')
df2
```

```python
# or 
df2 = df1.query('국어 < 70 or 수학 < 70')
df2
```

```python
# 변수 사용
a1 = 90

df2 = df1.query('국어 > @a1') # 변수 a1
df2
```

```python
list1 = ['철수', '영희', '민수']

# 리스트 안에 포함되어 있는 행만 가져온다.
df2 = df1.query('이름 in @list1')
df2
```

```python
# 리스트 안에 포함되지 않는 것만 가져온다.
df2 = df1.query('이름 not in @list1')
df2

#조건문이 더 직관적
```

### 행 추가

```python
df1 = pd.read_csv('data/grade.csv', encoding = 'euc-kr')
df1.set_index('이름', inplace=True)
df1
```

```python
# 철수의 모든 정보를 변경한다.
# 컬럼의 개수만큼 데이터를 넣어준다.

df1.loc['철수'] = [5, '외계인', 2000, 3000, 4000, 5000]
df1
```

```python
# 특정 컬럼의 모든 값을 변경한다.

df1['과학'] = [100, 200, 300, 400, 500]
df1
```

```python
# 특정 행의 일부만 변경한다.

df1.loc['호영', ['국어', '영어']] = 1000, 2000
df1
```

```python
# 특정 컬럼의 일부 행만 변경한다.
df1.loc[['민수', '수현'], '국어'] = [500, 600]
df1
```

```python
# 일부 행, 일부 열의 값을 변경한다.

df1.loc[['영희', '민수'], ['수학', '과학']] = [[100, 200], [300, 400]]
df1
```

```python
# 행 추가
# 존재하지 않는 인덱스로 값을 설정해주면 행이 추가된다.
# 있는 이름은 덮어씌어지고 없는 이름에는 새로 값이 추가된다.

df1.loc['영민'] = [5, '여자', 1, 2, 3, 4]
df1
```

```python
# 열 추가
# 없는 이름의 컬럼을 사용하여 값을 설정하면 추가된다.

df1['한국사'] = [1,2,3,4,5,6]
df1
```

```python
# 일부에만 지정한다.
a1 = {
    '국어' : 100,
    '영어' : 300
}

df1.loc['국영수'] = a1
df1
```
