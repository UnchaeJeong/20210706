데이터 분석에 앞서 전처리가 완료된 데이터에 대해 결측치나 이상치를 처리

- 제거 : 결측치나 이상치를 가진 데이터를 제거하는 것을 의미한다.
- 제거를 하게 되면 데이터가 왜곡되지 않지만 학습량이 줄어든다.
- 대체 : 결측치나 이상치를 다른 값으로 바꾸는 것을 의미한다.
- 대체를 하게 되면 데이터가 왜곡되지만 학습량은 유지된다.
- 제거를 추천. 제거 후 성능이 매우 안좋다면 대체할 값을 찾아야 한다.
- 대체를 선택했을 때에는 누구나 납득할 수 있는 객관적인 지표를 찾아야하고, 찾을 수가 없다면 평균이나 중간값을 추천한다.


```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df1 = pd.read_csv('data/grade.csv', encoding='euc-kr')
df1.set_index('이름', inplace=True)
df1
```

### 결측치 확인


```python
# Non-Null Count : 결측치를 제외한 데이터의 개수
df1.info()
```

```python
# count는 결측치를 제외한 데이터의 개수를 알아올 수 있다.
df1.count()
```

```python
# isna : 결측치인 부분은  True,
##결측치가 아닌 부분을 False로 되어 있는 데이터
##프레임을 반환한다.
# sum : 각 컬럼의 총 합을 구한다.
##True나 False로 되어 있을 경우 각각 1, 0으로 환산하여 총합을 구한다.
df1.isna().sum()
```

### 결측치 제거

```python
# 결측치가 포함된 모든 행을 제거한다.
df2 = df1.dropna()
df2
```

```python
# 특정 컬럼에 결측치가 있는 로우를 제거한다.
df3 = df1.dropna(subset=['수학', '과학'])
df3
```

```python
# 모든 컬럼의 결측치가 있는 로우를 삭제한다.
# 지금까지 프로젝트 하면서 한 번도 써본 적이 없다.ㅋ

df1.loc['길동'] = [np.nan,np.nan,np.nan,np.nan,np.nan,np.nan]
display(df1)

df1.dropna(how='all', inplace=True)
df1
```

### 결측치가 포함된 열 제거

```python
# 결측치 비율이 너무 많은 열의 경우 제거하는 게 나을 수 있다.
# 결측치가 하나라도 들어있는 컬럼은 모두 제거한다.

df2 = df1.dropna(axis=1)
df2
```

### 결측치 대체

```python
# 데이터프레임의 모든 결측치를 50으로 설정한다.

df2 = df1.fillna(value=50)
df2
```

```python
# 컬럼별로 결측치를 대체한다.
df1['영어'].fillna(value=50, inplace=True)
df1['수학'].fillna(value=60, inplace=True)
df1['과학'].fillna(value=70, inplace=True)
df1
```
