```python
corr = train.corr()
mask = np.zeros_like(corr, dtype=np.bool)
# 반만 나오게
mask[np.triu_indices_from(mask)] = True
plt.figure(figsize=(14,8))
plt.title('Overall Correlation of Titanic Features', fontsize=18)
sns.heatmap(corr, mask=mask, annot=True, cmap='RdYlGn', linewidths=0.2,
            annot_kws={'size':25})
# annot은 안에 값 나오는 함수. annot_kws={size}와 같이 외울 것.
plt.show()
```

# Part 2 'Survived'

## 생존자 숫자 보기
### sns.countplot(y='',data=train) => barh or (x='', data=train) => bar

```python
# fig = plt.figure(figsize=(10,2))
sns.countplot(y='Survived', data=train)
print(train.Survived.value_counts())
```

```python
# fig = plt.figure(figsize = (10, 5))
sns.countplot(x='Survived', data=train)
print(train['Survived'].value_counts())
```

## 파이차트 + 카운트플롯
### 1) f,ax  2) train[''].value_counts().plot.pie()

```python
f,ax=plt.subplots(1,2,figsize=(15,6))
train['Survived'].value_counts().plot.pie(explode=[0,0.1],autopct
                                         ='%1.1f%%', ax=ax[0], shadow=True)
ax[0].set_title('Survived')
ax[0].set_ylabel('')
sns.countplot('Survived', data=train, ax=ax[1])
ax[1].set_title('Survived')
plt.show()
```
