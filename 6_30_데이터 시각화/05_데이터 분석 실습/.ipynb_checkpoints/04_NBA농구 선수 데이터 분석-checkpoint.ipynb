```python
# 데이터 프레임 정보 확인
df1.info()
```

### 데이터 전처리

```python
# 컬럼 이름을 확인한다.
df1.columns
# 사용하기 편한 이름으로 컬럼이름 정하세요.
```

```python
# 컬럼 이름을 변경한다.
df1.columns = ['경기일자', '경기시간', '방문팀', '방문팀점수', '홈팀', '홈팀점수', 'Box','연장전여부', '관중수', 'Notes']
df1
```

```python
# 결측치 확인
df1.isna().sum()
```

```python
# 연장전 여부에 저장되어 있는 값의 종류를 확인한다.
df1['연장전여부'].value_counts()
```

```python
# 연장전 여부에 결측은 연장전을 하지 않은 경기이다.
# 따라서 결측치를  NOT으로 설정한다.

df1['연장전여부'].fillna('NOT', inplace=True)
df1['연장전여부'].value_counts()
```

```python
df1.isna().sum()
```

```python
# 날짜 컬럼의 타입을 확인한다.
df1['경기일자']
```

```python
# object를 날짜데이터로 바꿔줄 것. 흐름으로 볼 수 있기 떄문.
# 날짜 데이터는 원하는 양식으로 변경하는 것이 좋다.
# %a : 요일 .......https://docs.python.org/3/library/datetime.html#strftime-and-strptime-behavior
# Tue, Oct 17, 2017
df1['경기일자'] = pd.to_datetime(df1['경기일자'], format='%a, %b %d, %Y')
df1['경기일자']
```

<img src='images/date.png' style="width:600px;"/>

```python
# Note, Box 제거
df1_na = df1.drop(['Notes', 'Box'], axis=1)
df1_na
```

### 필요한 컬럼만 추출해서 정리한다.

```python
a1 = ['경기일자', '방문팀', '방문팀점수', '홈팀','홈팀점수']
games_nae = df1_na[a1]
games_nae
```

```python
# 경기 일자 컬럼을 인덱스로 지정한다.
# append에 True를 넣어주면 이전에 있던 index도 유지된다.
games_naes = games_nae.set_index('경기일자', append=True)
games_naes
```

```python
# 인덱스의 이름을 설정한다.
games_naesr = games_naes.rename_axis(['게임', '경기일자'])
games_naesr
```

### melt함수 사용해보기

```python
df = pd.DataFrame({
    'A' : ['a1', 'a2', 'a3', 'a4'],
    'B' : ['b1', 'b2', 'b3', 'b4'],
    'C' : [1,2,3,4],
    'D' : [100, 200, 300, 400]
})
df
```

```python
df.melt(id_vars='A', value_vars=['B','C'])
```

```python
# 인덱스를 다시 설정한다.
a1 = games_naesr.reset_index()
a1
```

```python
tidy = pd.melt(a1, id_vars=['게임', '경기일자'],
              value_vars=['방문팀', '홈팀'])
tidy
```
