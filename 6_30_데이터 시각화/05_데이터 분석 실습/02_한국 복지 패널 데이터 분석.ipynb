### 성별에 따른 평균 월급 차이 분석

```python
# 데이터 추출
df2 = df[['h12_g3', 'p1202_8aq1']]
df2
```

```python
# 컬럼 이름 변경
df2.columns = ['성별', '월급']
df2
```

```python
# 성별 데이터를 변경한다.
a1 = {
    1 : '남자',
    2 : '여자'
}
df2['성별'] = df2['성별'].map(a1)
df2
```

```python
#결측치 확인
df2.isna().sum()
```

```python
# 이상치 확인
df2['월급'].agg(['min','max'])
```

```python
# 결측치인 로우를 모두 제거한다.
df2.dropna(inplace=True)
df2.isna().sum()
```

```python
# 성별에 따른 월급 평균을 구한다.
# 성별과 급여평균 형태로 되어 있는 데이터 프레임을 생성한다.
a1 = df2.groupby('성별').mean()
display(a1)
```

```python
# index를 데이터에 포함시킨다.
a1.reset_index(inplace=True)
a1
```

```python
#시각화
plt.figure(figsize=(12,12))
ax = sns.barplot(x='성별', y='월급', data=a1)

for idx, value in enumerate(a1['월급']) :
    #표시할 문자열
    txt = f'{int(value)}만원'
    #문자열을 표시한다
    plt.text(idx, value + 10, txt, color='red',
            horizontalalignment = 'center')
    
plt.ylim(0,500)
    
plt.show()
```

### 나이에 따른 평균 월급의 변화

```python
# 데이터 추출
df2 = df[['h12_g4', 'p1202_8aq1']]
df2
```

```python
# 컬럼 이름을 변경한다.
df2.columns = ['태어난 연도', '월급']
df2
```

```python
# 나이를 구해 추가해준다.
df2['나이'] = 2017 - df2['태어난 연도'] + 1
df2
```

```python
# 결측치 확인
df2.isna().sum()
```

```python
# 이상치 확인
df2.agg(['min', 'max'])
```

```python
# 결측치 처리(채워보기)
# 월급이 결측치인 곳은 평균으로 채워준다.
# 월급의 평균을 구한다.
m1 = df2['월급'].mean()
print(f'월급 평균 : {int(m1)}')
#절대 이렇게 하면 안됨. 노동부 자료를 토대로 세대별로 평균값을 채워줄 것.
```

```python
# 월급 평균 257이기 때문에 제 마음대로 정하겠습니다.
# 10살 미만 : 0원
# 10~ 20살 : 30만원
# 20~ 30살 : 200만원
# 30~ 40살 : 300만원
# 40~ 50살 : 400만원
# 50살 이상 : 300만원

# df2.isna()

# df2['월급'].isna()

# (df2['월급'].isna()) & (df2['나이'] < 10)

# df2[(df2['월급'].isna()) & (df2['나이'] < 10)]

# df2[(df2['월급'].isna()) & (df2['나이'] < 10)].index

# idx1 = df2[(df2['월급'].isna()) & (df2['나이'] < 10)].index

# idx1 = df2[(df2['월급'].isna()) & (df2['나이'] < 10)].index
# idx2 = df2[(df2['월급'].isna()) & (df2['나이'] >= 10)].index
# idx3 = df2[(df2['월급'].isna()) & (df2['나이'] >= 20)].index
# idx4 = df2[(df2['월급'].isna()) & (df2['나이'] >= 30)].index
# idx5 = df2[(df2['월급'].isna()) & (df2['나이'] >= 40)].index
# idx6 = df2[(df2['월급'].isna()) & (df2['나이'] >= 50)].index


idx1 = df2[(df2['월급'].isna()) & (df2['나이'] < 10)].index
idx2 = df2[(df2['월급'].isna()) & (df2['나이'] >= 10) & (df2['나이'] < 20)].index
idx3 = df2[(df2['월급'].isna()) & (df2['나이'] >= 20) & (df2['나이'] < 30)].index
idx4 = df2[(df2['월급'].isna()) & (df2['나이'] >= 30) & (df2['나이'] < 40)].index
idx5 = df2[(df2['월급'].isna()) & (df2['나이'] >= 40) & (df2['나이'] < 50)].index
idx6 = df2[(df2['월급'].isna()) & (df2['나이'] >= 50)].index

df2.loc[idx1, '월급'] = 0
df2.loc[idx2, '월급'] = 30
df2.loc[idx3, '월급'] = 200
df2.loc[idx4, '월급'] = 300
df2.loc[idx5, '월급'] = 400
df2.loc[idx6, '월급'] = 300
           
df2['월급'].isna().sum()
```

```python
# 나이별 급여 평균을 구한다.
df3 = df2[['월급', '나이']]

result_df = df3.groupby('나이').mean()
result_df
```

```python
result_df.plot()
plt.show()
```

### 성별과 연령대 분포

```python
df2 = df[['h12_g3', 'h12_g4']]
df2.columns = ['성별','태어난 연도']
df2['나이'] = 2017 - df2['태어난 연도'] + 1
df2.drop('태어난 연도', axis=1, inplace=True)
df2
```

```python
# 성별 변경
a1 = {
    1 : '남자',
    2 : '여자'
}

df2['성별'] = df2['성별'].map(a1)
df2
```

```python
# 연령대를 구한다.
df2['연령대'] = df2['나이'] // 10 * 10
df2
```

```python
# 연령대별 성별 수를 구한다.

# df2.groupby(['연령대', '성별']).count()
result_df = df2.groupby(['연령대', '성별'], as_index=False).count()
result_df
# as_index? 인덱스가 생겨남. 와우
```

```python
result_df.columns = ['연령대', '성별', '명']

sns.barplot(x='연령대', y='명', hue='성별', data = result_df)
plt.show()
```
