```python
import matplotlib.pyplot as plt
import numpy as np

#그래프를 그리기 위한 기본 설정
#폰트 설정 - 윈도우
plt.rcParams['font.family'] = 'Malgun Gothic'
# 폰트 설정 - 맥
# plt.rcParams['font.family'] = 'AppleGothic'

#그래프의 크기
plt.rcParams['figure.figsize'] = 12, 6

#글자 크기
plt.rcParams['font.size'] = 14

#폰트 설정 시 - 기호 깨지는 것 방지하기
plt.rcParams['axes.unicode_minus'] = False

plt.rcParams['figure.facecolor'] = 'white'
```

### 선 그래프
- 어떠한 데이터 흐름에 따라 변화하는 데이터를 표현할 때 사용한다. 

```python
X = [1, 2, 3]
Y = [2, 3, 1]
plt.title('선 그래프')
plt.plot(X,Y)

# 각 축의 값의 종류(숫자, 문자..)와 동일한 것으로 설정해야 한다.
# 선 그래프에 어울리는 데이터는 숫자인 경우이기 때문에 숫자로 설정하면 된다.
plt.xticks([1,2,3], ['가', '나', '다'])
plt.yticks([0,2,4], ['하', '중', '상'])

#각 축의 이름
plt.xlabel('x축')
plt.ylabel('y축')

plt.show()
```

### 막대 그래프
- 데이터들의 양을 비교하기 위해 사용한다.


```python
X = [1, 2, 3]
Y = [2, 3, 1]
plt.title('막대 그래프')
plt.bar(X,Y)

# 각 축의 값의 종류(숫자, 문자..)와 동일한 것으로 설정해야 한다.
# 선 그래프에 어울리는 데이터는 숫자인 경우이기 때문에 숫자로 설정하면 된다.
plt.xticks([1,2,3], ['가', '나', '다'])
plt.yticks([0,2,4], ['하', '중', '상'])

#각 축의 이름
plt.xlabel('x축')
plt.ylabel('y축')

plt.show()
```

```python
X = [1, 2, 3]
Y = [2, 3, 1]
plt.title('가로 방향 막대 그래프')
plt.barh(Y,X)

# 각 축의 값의 종류(숫자, 문자..)와 동일한 것으로 설정해야 한다.
# 선 그래프에 어울리는 데이터는 숫자인 경우이기 때문에 숫자로 설정하면 된다.
plt.yticks([1,2,3], ['가', '나', '다'])
plt.xticks([0,2,4], ['하', '중', '상'])

#각 축의 이름
plt.xlabel('x축')
plt.ylabel('y축')

plt.show()
```

### 스템 플롯
- 막대 그래프와 유사하지만 폭이 없는 그래프이다.
- 이산 확률 함수나 자기 상관 관계를 묘사할 때 사용한다.

```python
x = np.linspace(0.1, 2*np.pi, 10)
y = np.cos(x)
y

plt.title('스템 플롯')
plt.stem(x, y, '-.')
plt.show()
```

### 파이 차트
- 카테고리 별 값의 상대적인 비교를 해야 할 때
- 막대 그래프는 양을 비교하는 것이고 파이차트는 비율을 비교한다.


```python
#알아서 백분율로 환산해줌.

values = [15, 30, 40, 10]
labels = ['값1', '값2', '값3', '값4']
colors = ['yellow', 'deepskyblue', 'slateblue', 'snow']
explode = [0,0.2,0,0]
#첫번째 : 값, labels : 파이 조각의 이름, colors : 파이의 색상
#explode : 파이의 튀어나옴 정도, shadow : 그림자 표시 여부
# startangle : 원의 시작 각도 값(기본이 0도는 3시 방향, 시계 반대 방향)
# counterclock : 원의 방향을 시계 반대 방향으로 할 것인가, 기본이 True
# autopct : 파이에 설정한 비율 값 양식, 설정하지 않으면 나타나지 않는다.

plt.title('파이 차트')
plt.pie(values, labels=labels, colors=colors, explode=explode,
       shadow=True, startangle = 90, counterclock=False,
       autopct='%.2f%%')
plt.show()
```
