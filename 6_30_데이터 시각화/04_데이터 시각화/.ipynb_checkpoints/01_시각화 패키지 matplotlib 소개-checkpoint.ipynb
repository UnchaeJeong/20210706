- Matplotlib는 파이썬에서 자료를 시각화하는 패키지
- 라인 플롯, 스캐터 플롯, 컨투어 플롯, 서피트 플롯, 
- 바 차트, 히스토그램, 박스 플롯 등등 다양한 그래프를 그릴 수 있다.
- https://matplotlib.org/stable/gallery/index.html

```python
import matplotlib.pyplot as plt
import matplotlib as mpl

#그래프를 그리기 위한 기본 설정
#폰트 설정 - 윈도우
plt.rcParams['font.family'] = 'Malgun Gothic'
# 폰트 설정 - 맥
# plt.rcParams['font.family'] = 'AppleGothic'

#그래프의 크기
plt.rcParams['figure.figsize'] = 12, 6

#글자 크기
plt.rcParams['font.size'] = 14

#폰트 설정 시 - 기호 깨지는 것 방지하기
plt.rcParams['axes.unicode_minus'] = False
```

### 라인 플롯
- 가장 간단한 그래프이다.
- 데이터가 시간, 순서 등에 따라 어떻게 변화되는지를 보여주기 위해 사용한다.
- (시계열 데이터, 값의 변화에 따른 특정 데이터 흐름...)

```python
plt.title('선 그래프')
# y 축에 설정될 값만 설정하면 x 축은 값의 개수로 설정된다.
plt.plot([1, 4, 9 , 16])
plt.show()
# plt.show()를 뒤에 붙여주면 실행. 그리고 각각 그래프를 산출.
# 원래는 하나의 그래프 안에 표현.
```

```python
plt.title('선 그래프')
plt.plot([-3, 4, 9, 16])
plt.show()
```

```python
plt.title('선 그래프')
# 리스트를 두 개 설정하면 x축, y축이 된다.
plt.plot([10,20,30,40],[1,4,9,16])
plt.show()
```

### 각 축의 이름 설정

```python
plt.title('선 그래프')
# 리스트를 두 개 설정하면 x축, y축이 된다.
plt.plot([10,20,30,40],[1,4,9,16])

# x 축의 이름
plt.xlabel('x축')
plt.ylabel('y축')

plt.show()
```

### 제목과 각 축의 글자 설정을 다르게 한다.

```python
option1 = {
    'family' : 'NanumMyeongjo', 'size' : 24, 'color' : 'black'
}
option2 = {
    'family' : 'NanumMeyongjo', 'size' : 18, 'color' : 'darkred'
}
option3 = {
    'family' : 'NanumBarumGothic', 'size' : 12, 'color' : 'blue'
}



plt.title('선 그래프', fontdict = option1)
# 리스트를 두 개 설정하면 x축, y축이 된다.
plt.plot([10,20,30,40],[1,4,9,16])

# x 축의 이름
plt.xlabel('x축', fontdict=option2)
plt.ylabel('y축', fontdict=option3)

plt.show()
```
