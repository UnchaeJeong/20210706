```python
fmri = sns.load_dataset('fmri')
fmri
```

```python
sns.relplot(x='timepoint', y='signal', kind='line', data=fmri);
```

```python
sns.relplot(x='timepoint', y='signal', sort=False, kind='line', data=fmri);
```

```python
sns.relplot(x='timepoint', y='signal', style='region', size='event', kind='line', data=fmri);
```

```python
sns.relplot(x='timepoint', y='signal', hue='subject', style='region', size='event', kind='line', data=fmri);
```

```python
sns.relplot(x='timepoint', y='signal', hue='subject', col='region', size='event', kind='line', data=fmri);
```

```python
palette = sns.cubehelix_palette(n_colors=14, light=0.8)
sns.relplot(x='timepoint', y='signal', hue='subject', col='region', size='event', kind='line', palette=palette, data=fmri);
```

```python
palette = sns.cubehelix_palette(n_colors=14, light=0.8)
sns.relplot(x='timepoint', y='signal', hue='subject', col='region', style='event', kind='line', palette=palette, data=fmri);
```

```python
palette = sns.cubehelix_palette(n_colors=14, light=0.8)
sns.relplot(x='timepoint', y='signal', hue='subject', col='region', kind='line', palette=palette, data=fmri.query("event == 'cue'"));
```

```python
sns.relplot(x='timepoint', y='signal', hue='subject', col='region', row='event', kind='line', palette=palette, data=fmri);
```

```python
sns.relplot(x='timepoint', y='signal', hue='event', col='subject',style='event', col_wrap=4, linewidth=3,
            kind='line', data=fmri.query("region == 'parietal'"));
```
