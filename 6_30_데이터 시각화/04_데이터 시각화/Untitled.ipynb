```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

#그래프를 그리기 위한 기본 설정
#폰트 설정 - 윈도우
plt.rcParams['font.family'] = 'Malgun Gothic'
# 폰트 설정 - 맥
# plt.rcParams['font.family'] = 'AppleGothic'

#그래프의 크기
plt.rcParams['figure.figsize'] = 12, 6

#글자 크기
plt.rcParams['font.size'] = 14

#폰트 설정 시 - 기호 깨지는 것 방지하기
plt.rcParams['axes.unicode_minus'] = False

plt.rcParams['figure.facecolor'] = 'white'
```

```python
iris = pd.read_csv('data/iris.csv')
iris.columns = ['꽃받침길이', '꽃받침너비', '꽃잎길이', '꽃잎너비', '품종']
iris
```

```python
sns.rugplot(iris['꽃잎길이'])
plt.title('꽃잎 길이에 대한 rug plot')
plt.show()
```

```python
sns.kdeplot(iris['꽃잎길이'])
plt.title('kde')
plt.show()
```

```python
iris
```

```python
sns.distplot(iris['꽃잎길이'], kde=True, rug=True)
plt.title('distplot')
plt.show()
```

```python
sns.countplot(data=iris, x='품종')
plt.title('count plot')
plt.show()
```

```python
sns.jointplot(x='꽃받침길이', y = '꽃받침너비', data = iris)
plt.suptitle('joint', y = 1.02)
plt.show()
```

```python
sns.jointplot(x='꽃받침길이', y = '꽃받침너비', kind = 'kde', data = iris)
plt.suptitle('joint', y = 1.02)
plt.show()
```

```python
sns.pairplot(iris)
plt.suptitle('pair', y = 1.02)
plt.show()
```

```python
sns.pairplot(iris, hue = '품종')
plt.suptitle('pairplot', y=1.02)
plt.show()
```

```python
mpg_df = pd.read_excel('data/mpg.xlsx')
mpg_df
```

```python
a1, a2 = mpg_df.shape
print(f'행의 크기 : {a1}')
print(f'열의 크기 : {a2}')
```

```python
mpg_df.info()
```

```python
df = mpg_df
df
```

```python
mpg_a = df.query('displ <= 4')
a_mean = mpg_a['hwy'].mean()
a_mean
```

```python
mpg_b = df.query('displ > 4')
b_mean = mpg_b['hwy'].mean()
b_mean
```

```python
d1 = {
    '배기량 4이하' : [a_mean],
    '배기량 4초과' : [b_mean]
}
print(d1)
tmp_df = pd.DataFrame(d1)
print()
tmp_df.index = ['고속도로 연비 평균']
display(tmp_df)
print()
tmp_df.plot.bar()
plt.show()

```

```python
d1 = {
    '고속도로 평균 연비' : [a_mean, b_mean]
}
print(d1)
print()
temp_df = pd.DataFrame(d1)
print(temp_df)
print()
temp_df.index = ['배기량 4이하', '배기량 4초과']
display(temp_df)
print()
temp_df.plot.bar()
plt.show()
```

```python
d1 = {
    '배기량' : ['4이하', '4초과'],
    '고속도로 평균연비' : [a_mean, b_mean]
}
temp_df = pd.DataFrame(d1)
display(temp_df)

sns.barplot(x='배기량', y='고속도로 평균연비', data=temp_df)
plt.show()
```

```python
df_tmp = mpg_df[['manufacturer', 'cty']]
df_tmp
```

```python
sns.barplot(x='제조사', y='시내 평균 연비', data=df_result)
plt.show()
```

```python

sns.countplot(x='성별',data=df2)

plt.title('조사 대상에 대한 성별 분포')

plt.ylabel('명')


a100 = df2['성별'].value_counts()

for idx, value in enumerate(a100) :
    txt = f'{value}명'
    plt.text(idx, value + 100, txt)
#             ,horizontalalignment = 'center')


plt.show()
```
