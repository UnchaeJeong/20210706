```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set(style='whitegrid')

%matplotlib inline
```

```python
tips = sns.load_dataset('tips')
tips
```

```python
sns.relplot(x='total_bill', y='tip', data=tips)
```

```python
sns.relplot(x='total_bill', y='tip', data=tips, hue='smoker')
```

```python
sns.relplot(x='total_bill', y='tip', data=tips, hue='size')
```

```python
sns.relplot(x='total_bill', y='tip', data=tips, style='smoker', hue='size')
```

```python
sns.relplot(x='total_bill', y='tip', data=tips, hue='size', palette ='terrain')
```

```python
sns.relplot(x='total_bill', y='tip', data=tips, size='size', palette ='terrain')
```

```python
import numpy as np
import pandas as pd
import matplotlib as mpl
import matplotlib.pyplot as plt
import seaborn as sns
sns.set(style="whitegrid")
```

```python
penguins = sns.load_dataset("penguins")
penguins
```

```python
sns.relplot(x="bill_length_mm", y="bill_depth_mm", data=penguins);
```

```python
sns.relplot(x="bill_length_mm", y="body_mass_g", hue = "bill_length_mm", data=penguins);
```

```python
sns.relplot(x="bill_length_mm", y="body_mass_g", hue = "species", style="species", data=penguins);
```

```python
sns.relplot(x="bill_length_mm", y="body_mass_g", hue = "species", style="island", data=penguins);
```

```python
sns.relplot(x="bill_length_mm", y="body_mass_g", hue = "species", col="island", data=penguins);
```

```python
sns.relplot(x="bill_length_mm", y="body_mass_g", hue = "body_mass_g", style="island", data=penguins);
```

```python
sns.relplot(x="bill_length_mm", y="body_mass_g", hue = "body_mass_g", size='body_mass_g', style="island", data=penguins);
```

```python
flights = sns.load_dataset("flights")
flights
```

```python
sns.relplot(x = 'year', y = 'passengers', kind='line', data=flights)
```

```python
dots = sns.load_dataset('dots')
dots
```

```python
sns.relplot(x='time', y='firing_rate',
           kind='line', data=dots);
```

```python
sns.relplot(x='time', y='firing_rate',
           ci = None, kind='line', data=dots);
```

```python
sns.relplot(x='time', y='firing_rate',
           ci ='sd', kind='line', data=dots);
```

```python
sns.relplot(x='time', y='firing_rate',
           estimator=None, kind='line', data=dots);
```

```python
sns.relplot(x='time', y='firing_rate',
           hue='choice', kind='line', data=dots);
```

```python
sns.relplot(x='time', y='firing_rate',
           hue='align', style = 'choice', kind='line', data=dots);
```
