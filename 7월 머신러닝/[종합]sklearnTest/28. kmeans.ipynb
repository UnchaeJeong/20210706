```python
from sklearn.cluster import KMeans 
import numpy as np 
import pandas as pd 
import seaborn as sb 
import matplotlib.pyplot as plt
import mglearn
```

```python
mglearn.plots.plot_kmeans_algorithm()
```

```python
df = pd.read_csv( 'data5/kmean.csv', index_col='Unnamed: 0')
df
```

```python
plt.scatter( df['x'], df['y'])
plt.show()
```

## 비지도학습( 군집분류)

```python
k = KMeans( n_clusters= 4 )
km = k.fit( df )
```

```python
type(km)
```

```python
cen = km.cluster_centers_
cen
```

```python
plt.scatter( df['x'], df['y'])
plt.scatter( cen[:, 0], cen[:, 1], marker= 's', c= ['r','g','b','y'])
plt.show()
```

```python
km.labels_
```

```python
df['labels'] = km.labels_
```

```python
df
```

```python
sb.pairplot( df, hue= 'labels')
plt.show()
```

```python
sb.lmplot(x='x', y='y', data= df, hue='labels', fit_reg=False, scatter_kws={'s':150})
plt.show()
```

### km.inertia_  각센터를 중심으로 떨어진 거리의 총합의 평균

```python
data = []
for n in range(1,15):
    model = KMeans( n_clusters=n)
    km = model.fit( df )
    data.append( km.inertia_ )
```

```python
plt.plot( range(1,15) ,data , 'ro--')
plt.show()
```
