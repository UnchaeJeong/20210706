# 1번

```python
df=pd.read_csv('data/electric.csv')
df
```

```python
df = df.drop(columns=['Unnamed: 0'], axis=1)
df
```

```python
x_data=df[['전기생산량']]
y_data=df['전기사용량']
```

```python
model_elec=LinearRegression()
model_elec.fit(x_data,y_data)
```

```python
a=model_elec.coef_
a
```

```python
b=model_elec.intercept_
b
```

```python
model_elec.predict([[3.2],[4.5]])
```

```python
y_pred=b+a*x_data
```

```python
plt.scatter(x_data,y_data)
plt.plot(x_data,y_pred,c='r')
plt.xlabel('전기생산량')
plt.ylabel('전기사용량')
plt.show()
```

# 2번
2번은 잘모르겠습니다 😥

```python
x=x_data
y=y_data
```

```python
w=1
b=1
n=len(x)
epochs=1000
learning_rate=0.01
```

```python
for i in range(epochs):
    hy=w*x+b
    cost=np.sum((hy-y)**2)/n
    gradientW=np.sum((w*x+b-y)*2*x)/n
    gradientB=np.sum((w*x+b-y)*2)/n
    w=w-learning_rate*gradientW
    b=b-learning_rate*gradientB
    print('cost:',cost,'w:',w,'b:',b)
print('최종w:',w) #2.9 ==> 3
print('최종b:',b) #b ==> 2
```

### (3x+2)^3
- 3x+2=t
- t^3을 t에 대해 미분 ==> 3t^2
- (3x+2)를 x에 대해 미분 ==> 3
- 두개를 곱한다. 3t^2 * 3
- 3(3x+2)^2 * 3 = 9(3x+2)^2
