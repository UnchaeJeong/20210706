###  1번
electric.csv  w, b를 구하시고
전기생산량이 3.2, 4.5 인경우 전기 사용량을 예측하시요.
scatter(실제데이터), plot(예측데이터)를 그리시요.

```python
etcDF = pd.read_csv('./data/electric.csv')
e_pro = etcDF[['전기생산량']]
e_con = etcDF[['전기사용량']]
```

```python
modelETC = LinearRegression()
modelETC.fit(e_pro,e_con)
x = np.array([[3.2], [4.5]])
y_predict = modelETC.predict(x)

plt.xlabel('전기생산량')
plt.ylabel('전기사용량')
plt.scatter(e_pro,e_con)
plt.plot(x,y_predict,'r')
plt.show()
```

```python
print('w :', modelETC.coef_,'b :', modelETC.intercept_)
```

### 2번
선형회귀 통계공식을 파이썬코드로 직접 구현해서
1 번데이터 w,b를 구하시요

```python
def cost(x,y,w):
    hy = w*x
    print(x, y, w)
    cost = np.sum((hy-y)**2)
    print(cost)
    return cost/len(x)
```

```python
x = np.array(e_pro['전기생산량'])
y = np.array(e_con['전기사용량'])
cost(x, y, -1)
```

```python
lr = 0.02
w = 5
b = 5
n = len(x)
epochs = 1000
```

```python
for i in range(epochs):
    hy = w*x + b
    cost = np.sum((hy - y)**2)/n
    
    gradientW = np.sum(2*(w*x + b - y)*x)/n     
    gradientB = np.sum(2*(w*x + b - y))       
    w = w - lr*gradientW
    b = b - lr*gradientB
    print(cost, w, b)
```
