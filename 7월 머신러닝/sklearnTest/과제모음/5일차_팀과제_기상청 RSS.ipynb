```python
from bs4 import BeautifulSoup

import urllib.request as REQ
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

from datetime import datetime
from sklearn.preprocessing import OneHotEncoder, LabelEncoder
from sklearn.preprocessing import StandardScaler
from sklearn.neural_network import MLPRegressor
from sklearn.model_selection import train_test_split
from sklearn.model_selection import GridSearchCV
from sklearn.model_selection import cross_val_score
from sklearn.linear_model import Ridge,Lasso,ElasticNet, LinearRegression, SGDRegressor
from sklearn.preprocessing import PolynomialFeatures
from sklearn.pipeline import make_pipeline, Pipeline
from matplotlib import rc

import matplotlib
import seaborn as sns
rc('font', family='AppleGothic')
plt.rcParams['axes.unicode_minus'] = False
```

```python
kurl ='https://www.weather.go.kr/weather/forecast/mid-term-rss3.jsp?stnId=109'
kresponse = REQ.urlopen( kurl )
```

```python
# html parsing
ksoup = BeautifulSoup( kresponse, 'html.parser') #html.parser , html5lib
ksoup
```

```python
# 도시만 불러오기
for location in ksoup.select('location'):
    print(location.city.string)
```

```python
# 데이터 불러오기
for location in ksoup.select('location'):
    print(location.city.string)
    print("======================")
    for data in location.select('data'):
        print('날짜:',  data.tmef.string )
        print('최저 : ',data.tmn.text)
        print('최고 : ',data.tmx.text)
        print("-----------------------")
```

```python
# list로 type 변환
kList = []
for location in ksoup.select('location'):
    for data in location.select('data'):
        kList.append( {'도시':location.city.string,
                      '날짜':data.tmef.string,
                      '최저':int(data.tmn.string),
                      '최고':int(data.tmx.string)}  )
kList
```

---
데이터 전처리

```python
# DF로 type 변환
kdf = pd.DataFrame( kList)
kdf.head(2)
```

```python
# 열이름 변경
kdf = kdf.rename(columns={'도시':'cty','날짜':'date','최저':'tmin','최고':'tmax'})
kdf.head()
```

```python
# 열 타입 변경 (Series -> str -> datetime)
kdf['date'] = kdf['date'].map(str)
```

```python
kdf.info()
kdf.head(2)
```

```python
kdf['ttmin'] = kdf['tmin'].shift(-2)
kdf['ttmax'] = kdf['tmax'].shift(-2)
# print(kdf.head(10))
# kdf.info()
```

```python
kdf[['tmin','tmax','ttmin','ttmax']] = kdf[['tmin','tmax','ttmin','ttmax']].astype(np.float32)
kdf.info()
```

```python
# pd.set_option('display.max_row', 500)
# pd.set_option('display.max_columns', 100)
# kdf
```

```python
lst = [i % 12 != 11 for i in range(len(kdf))]
# kdf = kdf[lst]
kdf.reset_index(drop = True, inplace = True)
display(kdf[kdf['cty'] == '서울'])
display(kdf[kdf['cty'] == '여주'])
```

NaN 제거

```python
kdf = kdf.dropna()
```

```python
x_data = kdf.iloc[:,:-2]
y_data = kdf.iloc[:,4:]
```

```python
x_data = kdf[['cty','date','tmin','tmax']]
x_data
```

```python
x_data = pd.get_dummies(x_data)
x_data
```

```python
x_train, x_test, y_train, y_test = train_test_split(x_data, y_data, 
                                test_size=0.2,random_state=1)
model = make_pipeline(StandardScaler(), LinearRegression())
model.fit(x_train, y_train)
```

```python
model_ridge = make_pipeline(StandardScaler(), Ridge())
model_ridge.fit(x_train, y_train)
```

```python
display(model_ridge.score(x_train,y_train))
display(model_ridge.score(x_test,y_test))
```

```python
model_ridge5 = make_pipeline(StandardScaler(), Ridge(alpha=5.0))
model_ridge5.fit(x_train, y_train)
```

```python
display(model_ridge5.score(x_train,y_train))
display(model_ridge5.score(x_test,y_test))
```
