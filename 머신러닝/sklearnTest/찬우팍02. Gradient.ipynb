```python
import matplotlib.pyplot as plt
import  numpy as np
```

```python
def cost(x, y, w):
    c = 0
    for i in range(len(x)):    # x,y 입력값으로 하나씩 확인해보기위한 루프
        hx = w * x[i]          # y = ax + b 형태의 그래프
        c = c + (hx-y[i])**2   # 분산값의 누적값  => 분산값 + (예측치 - 실측치)**2
    return c/len(x)            # 누적값을 전체 데이터 수로 나누어 평균치.
```

```python
# 합성함수의 미분
# 기울기
def gradient(x, y, w):         # 기울기
    c = 0                      # 동일한 누적치
    for i in range(len(x)):    # x,y 입력값으로 하나씩 확인해보기위한 루프
        hx = w * x[i]          # y = ax + b 형태의 그래프
        c = c + (hx-y[i])*x[i] # 그래프의 미분값 결과를 통해 나온 수식을 그대로 함수에 대입.
    return c/len(x)            # 기울기값의 누적값 
```

```python
def show(x,y,w):              # 출력용
    hist = []                 # histogram
    for i in range(200):      # x,y 입력값으로 하나씩 확인해보기위한 루프
        c = cost(x,y,w)       # 코스트값(분산 평균)
        hist.append(c)        # 코스트 값을 히스토그램에 넣음 => 아마 나중에 출력하기위해서?
        print('i :', i , 'cost :', c, 'w :', w)
        g = gradient(x,y,w)   # 기울기 값
        w = w - 0.1 * g       # 기울기 값을 조정. 임의의 러닝레이트를 통해 너무 큰 hop이 아니도록 조정(수학의 좌극한, 우극한, 등의 개념과 유사)
#       # print(g)
        # 러닝레이트가 비교적 작은경우 최적값을 찾을 확률이 올라가지만, 연산 횟수, 시간이 증가,
        # 러닝레이트가 비교적 크다면, 근사치 혹은 대략적인 값을 금방 찾을 수는 있겠지만, 그러한 근사치중 최소값이 최적값이라는 보장이 없음.
```
