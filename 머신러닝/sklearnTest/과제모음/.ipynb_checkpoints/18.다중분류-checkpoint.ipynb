```python
iris_df = pd.DataFrame(iris['data'], columns=iris['feature_names'])
iris_df['specis'] = iris['target']
iris_df
```

```python
x_data = iris_df.iloc[:,:-1]
y_data = iris_df.iloc[:,-1]
```

```python
x_train, x_test, y_train, y_test = train_test_split(x_data, y_data,
                                                   test_size=0.2,
                                                   stratify=y_data)
```

```python
iris_df
# 굳이 정규화가 필요해 보이지 않음.
# 원래 파이프라인으로 정규화 진행해야함.
# 지금은 그냥 정규화 없이 진행해봄.
```

```python
# 다중분류 (// 이중분류)
model_logi = LogisticRegression(multi_class = 'multinomial')
# 'ovr' 2진법 sigmoid사용
# multinomial 다중분류 softmax를 사용
# multi_class='auto' -> 두개 중 자동으로 선택해준다라는 의미
# 이번엔 확실히 구분하기 위해 그냥 일부러 주겠음.
model_logi.fit(x_train, y_train)
```

```python
model_logi.score(x_test, y_test)
```

```python
model_logi.coef_
# setosa
# veriscala
# virginia
# 각각에 대한 feature 4개의 weight
```
