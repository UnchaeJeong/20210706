## 1.  1등급, female, 32달러 인경우 생사유무를 예측하시요

```python
x_data, y_data = titanic[['pclass', 'sex', 'fare']], titanic['survived']
```

```python
x_train, x_test, y_train, y_test = train_test_split(x_data, y_data, random_state=1, stratify=y_data)
x_train
```

```python
x_train = x_data[['pclass','sex','fare']]
x_test = x_data[['pclass','sex','fare']]
```

```python
pd.get_dummies(x_train, columns=['pclass','sex','fare'])
pd.get_dummies(x_test, columns=['pclass','sex','fare'])
```

```python
# y_train_ohe = ohe.fit_transform(y_train['survived'])
```

```python
#파이프라인
model_logi = make_pipeline(StandardScaler(), LogisticRegression())
model_knn = make_pipeline(StandardScaler(),KNeighborsClassifier())
model_tree = make_pipeline(StandardScaler(),DecisionTreeClassifier())
# vote
model_vote = VotingClassifier(estimators=[('logi',model_logi),
                                         ('knn',model_knn),
                                        ('tree',model_tree)],
                             voting='soft' )
# vote.fit
model_vote.fit(x_train, y_train)
# vote.score
# model_vote.score(x_test_ohe, y_test)
```
