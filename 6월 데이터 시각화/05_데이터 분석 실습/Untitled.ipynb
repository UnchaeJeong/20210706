```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_style('ticks')
sns.set_style('whitegrid')
plt.rcParams['font.family'] = 'Malgun Gothic'
plt.rcParams['figure.figsize'] = 12, 6
plt.rcParams['font.size'] = 14
plt.rcParams['axes.unicode_minus'] = False
plt.rcParams['figure.facecolor'] = 'white'
```

```python
mpg_b = mpg_df.query('displ > 4')
mpg_b
b1 = mpg_b['hwy'].mean()
b1
```

```python
d1 = {
    '배기량 4이하' : [a1],
    '배기량 4초과' : [b1]
}
d1_df = pd.DataFrame(d1)
d1_df.index = ['고속도로 연비 평균']
display(d1_df)

d1_df.plot.bar()
plt.show()
```

```python
d1 = {
    '배기량' : ['4이하', '4초과'],
    '고속도로 평균연비' : [a1, b1]
}
temp_df = pd.DataFrame(d1)
display(temp_df)
```

```python
sns.barplot(x='배기량',y='고속도로 평균연비', data=temp_df)
plt.show()
```

```python
df_tmp = mpg_df[['manufacturer','cty']]
df_tmp
```

```python
df_result = df_tmp.groupby('manufacturer').mean()
df_result
```

```python
display(df_result)
```

```python
# 도시 연비를 기준으로 정렬한다.
df_result.sort_values('cty', ascending=False, inplace=True)

df_result.columns = ['시내 평균 연비']

df_result.plot.bar()
plt.show()
```

```python
# 데이터프레임의 인덱스를 리셋한다.(이 때 drop=True를 하지 않으며
## 인덱스가 컬럼으로 포함된다.)
df_result.reset_index(inplace=True)
df_result.columns = ['제조사', '시내연비 평균']
df_result
```

```python
sns.barplot(x='제조사', y='시내연비 평균', data=df_result)
plt.show()
```
