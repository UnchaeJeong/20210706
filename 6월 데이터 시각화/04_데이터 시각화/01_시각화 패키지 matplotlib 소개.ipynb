## <span style="color:fuchsia">**üìö 8. Î≤îÎ°Ä**</span>


```python
import numpy as np
```

```python
#ÏãúÏûëÍ∞í, Ï¢ÖÎ£åÍ∞í, Ïà´ÏûêÍ∞úÏàò : ÏãúÏûëÍ∞íÏóêÏÑúÎ∂ÄÌÑ∞ Ï¢ÖÎ£åÍ∞íÍπåÏßÄ
##Ï¥ù Ïà´Ïûê Í∞úÏàòÎßåÌÅº Ïà´ÏûêÎ•º Í∞ÄÏßÄÍ≥† ÏûàÎäî ndarrayÎ•º ÏÉùÏÑ±ÌïúÎã§.

X = np.linspace(-np.pi, np.pi, 256)
y1 = np.cos(X)
y2 = np.sin(X)

plt.title('Î≤îÎ°Ä ÌëúÏãúÌïòÍ∏∞')
plt.plot(X, y1, ls='--', label='cosine')
plt.plot(X, y2, ls=':', label='sine')

#Í∞Å Í∑∏ÎûòÌîÑÏùò label Í∞íÏùÑ ÌôúÏö©ÌïòÏó¨ Î≤îÎ°ÄÎ•º ÌëúÏãúÌïúÎã§.
# loc : Î≤îÎ°ÄÏùò ÏúÑÏπò, ÏÉùÎûµ Ïãú ÏûêÎèô
# 0ÏùÄ (best, Í∏∞Î≥∏), 1(Ïö∞ÏÉÅÎã®), 2(Ïö∞Ï¢åÏ∏°), 3(ÏïÑÎûòÏ¢åÏ∏°), 4(ÏïÑÎûòÏö∞Ï∏°), 5(Ïö∞Ï∏°),
##6(Ï§ëÏïôÏ¢åÏ∏°), 7(Ï§ëÏïôÏö∞Ï∏°), 8(ÌïòÎã®Ï§ëÏïô), 9(ÏÉÅÎã®Ï§ëÏïô) 10(Ï†ïÏ§ëÏïô)
plt.legend(loc=10)
plt.show()
```
